# Dataset "Quantum Finance"

Este dataset contém informações sobre clientes de uma startup financeira, incluindo dados demográficos e financeiros que podem ser utilizados para análise de crédito e perfil dos clientes. As variáveis incluídas permitem aplicar técnicas de feature engineering e análise exploratória.

- Características do Dataset:
    - id: Identificador único do cliente.
    - idade: Idade do cliente (em anos).
    - sexo: Sexo biologico do cliente (masculino ou feminino).
    - estado_civil: Estado civil do cliente (solteiro, casado, divorciado).
    - escola: Grau de instrução do cliente (ensino médio, graduação, doutorado).
    - qte_dependentes: Número de dependentes do cliente.
    - tempo_ultimoservico: Tempo de trabalho do cliente (em meses).
    - trabalha: Cliente Trabalha? (sim ou nao).
    - vl_salario_mil: Valor do salario do cliente (em R$).
    - reg_moradia: Região da moradio do cliente (Segmento geografio de A a F).    
    - casa_propria: Cliente possui casa propria? (sim ou nao).
    - vl_imovel_em_mil: Valor do imovel do cliente (em R$).
    - qte_cartoes: Número de cartoes de credito do cliente.
    - qte_carros: Número de carros do cliente.
    - SCORE_CREDITO: Pontuação de crédito do cliente (em uma escala de 100 a 1000).

# 4 Redução de Dimensionalidade

## 4.1 Encoding Categorical

```python
# Pré-processamento     

# Validação de dados faltantes
print("\n### Dados Faltantes ###")  
#f_cleaned = df.replace('na', np.nan)     
missing_data = df.isnull().sum()
missing_percent = (df.isnull().sum() / len(df)) * 100
print(pd.DataFrame({'Missing Values': missing_data, 'Percentage': missing_percent}))

# Validação de registros duplicados   
print(f"\n### Total de registros duplicados: {df.duplicated().sum()} ###")    

# Remocao da coluna ID, pois não é relevante para análise
X = df.drop(columns=['id'])

# Identificando as variáveis categóricas
categorical_columns = X.select_dtypes(include=['object']).columns
numerical_columns = X.select_dtypes(include=['number']).columns

# Aplicando o encoding apenas às variáveis categóricas
X_categorical_encoded = pd.get_dummies(X[categorical_columns], drop_first=True).astype(int)

# Mantendo as variáveis numéricas como estão
X_numerical = X[numerical_columns]

# Concatenando variáveis categóricas codificadas e variáveis numéricas
X_encoded = pd.concat([X_numerical, X_categorical_encoded], axis=1) 
```

```python
X_encoded.head()
```

## 4.2 Normalização

```python
# Normalização
scaler = StandardScaler()
X_encoded_scaled = pd.DataFrame(scaler.fit_transform(X_encoded), columns=X_encoded.columns) 
```

```python
X_encoded_scaled.head()
```

## 4.3 PCA 

```python
# PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_encoded_scaled)
```

## 4.4 t-SNE

```python
# t-SNE
tsne = TSNE(n_components=2, random_state=0)
X_tsne = tsne.fit_transform(X_encoded_scaled)
```

## 4.5 LDA

```python
# LDA   
y = df['tempo_ultimoservico']
lda = LDA(n_components=2)
X_lda = lda.fit_transform(X_encoded_scaled, y)
```

## 4.6 Visualização

```python
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis')
plt.title('PCA')
#plt.xlabel('Componente Principal 1')
#plt.ylabel('Componente Principal 2')

plt.subplot(1, 3, 2)
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, cmap='viridis')
plt.title('t-SNE')
#plt.xlabel('Componente t-SNE 1')
#plt.ylabel('Componente t-SNE 2')

plt.subplot(1, 3, 3)
if X_lda.shape[1] > 1:
    plt.scatter(X_lda[:, 0], X_lda[:, 1], c=y, cmap='viridis')
    #plt.xlabel('Componente LDA 1')
    #plt.ylabel('Componente LDA 2')
else:
    plt.scatter(X_lda[:, 0], np.zeros_like(X_lda[:, 0]), c=y, cmap='viridis')
    #plt.xlabel('Componente LDA 1')
    #plt.ylabel('')

plt.suptitle('Visualização das Reduções de Dimensionalidade', fontsize=16)
plt.show()
```
